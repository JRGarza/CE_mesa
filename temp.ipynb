```python
# import sys
# sys.path.append('../python_scripts')

import matplotlib.pyplot as plt
import pylab as p
import numpy as np
import os
from glob import glob

import sys
sys.path.insert(0, '/data/disk1/fragkos/repos/CE_mesa/python_scripts')
import mesa as ms
import binary as bn


Lsun = 3.828e33
secyer = 24.0*3600.0*365.24
GGG = 6.674e-8
Msun = 1.989e33
Rsun = 6.955e10
clight = 29979245800
pi = 3.1415
crad = 7.57e-15



%matplotlib inline 
plt.rcParams['figure.figsize'] = 12, 6  # that's default image size for this interactive session

run_path= "/data/disk1/fragkos/mesa_projects/CE/DVU_Mexico/8M_30M_1400R/"
#run_path= "/data/disk1/fragkos/mesa_projects/CE/DVU_Mexico/1.4M_15M_700R/"


run_names = [d for d in os.listdir(run_path) if os.path.isdir(os.path.join(run_path, d))]

run_names = ["step2_lee_setE2_noMLT", "step2_macleod_noMLT", "step2_macleod_setE2", "step2_macleod_setE2_noMLT", "step2_macleod_setE2_noMLT_ML", "step2_macleod_setE2_noMLT_NoRot", "step2_macleod_setE2_noMLT_NoSync"]
```

```python
for run_name in run_names:
    data_path = run_path+run_name+"/LOGS/"
    a = ms.mesa(data_path=data_path, parallel=True, abundances=False, log_abundances = True, Yaxis='log_mass', Xaxis="model_number",czones=False, Variable='v_div_vesc', orbit=True)
    print len(a.profiles)

    fig, ax = plt.subplots(5,3, figsize=(36, 45))

#     masses_TML = np.linspace(0.5,13.5,27)
#     xvals_TML = np.linspace(1.1,1.9,27)
    
    masses_TML = np.linspace(0.5,27.,54)
    xvals_TML = np.linspace(1.1,1.9,54)


    a.SetParameters(cmap_dynamic_range=3.0, masses_TML=masses_TML, xvals_TML=xvals_TML,
                   signed_log_cmap=False, Yaxis='log_radius', tau10=True, tau100=True, Xaxis="star_age",
                   czones=False,cmap='jet', Variable='v_div_vesc')
    a.Kippenhahn(ax[0,0])

    a.SetParameters(cmap_dynamic_range=3.0, signed_log_cmap=False, Variable='eps_recombination')
    a.Kippenhahn(ax[0,1])


    a.SetParameters(cmap_dynamic_range=3.0, signed_log_cmap=False, Variable='v_div_csound')
    a.Kippenhahn(ax[0,2])

    
    a.SetParameters(cmap_dynamic_range=8.0, signed_log_cmap=False, Variable='opacity')
    a.Kippenhahn(ax[1,0])


    a.SetParameters(cmap_dynamic_range=5.0, signed_log_cmap=False, Variable='ionization_energy')
    a.Kippenhahn(ax[1,1])

    a.SetParameters(cmap_dynamic_range=5.0, signed_log_cmap=False, Variable='total_energy_plus_vconv2')
    a.Kippenhahn(ax[1,2])


    
    a.SetParameters(cmap_dynamic_range=10.0, signed_log_cmap=False, Variable='density')
    a.Kippenhahn(ax[2,0])

    a.SetParameters(cmap_dynamic_range=5.0, signed_log_cmap=False, Variable='Lrad_div_Ledd')
    a.Kippenhahn(ax[2,1])

    a.SetParameters(cmap_dynamic_range=5.0, signed_log_cmap=False, Variable='t_thermal')
    a.Kippenhahn(ax[2,2])

    
    
    a.SetParameters(cmap_dynamic_range=4.0, signed_log_cmap=False, Variable='total_energy')
    a.Kippenhahn(ax[3,0])

    a.SetParameters(cmap_dynamic_range=4.0, signed_log_cmap=False, Variable='j_rot')
    a.Kippenhahn(ax[3,1])

    a.SetParameters(cmap_dynamic_range=4.0, signed_log_cmap=False, Variable='extra_heat')
    a.Kippenhahn(ax[3,2])
    
    
    a.SetParameters(cmap_dynamic_range=4.0, signed_log_cmap=False, Variable='omega')
    a.Kippenhahn(ax[4,0])

    a.SetParameters(cmap_dynamic_range=4.0, signed_log_cmap=False, Variable='omega_div_omega_crit')
    a.Kippenhahn(ax[4,1])

    a.SetParameters(cmap_dynamic_range=4.0, signed_log_cmap=False, Variable='super_ad')
    a.Kippenhahn(ax[4,2])



    

    plt.tight_layout()
    plt.savefig("../figures/"+run_name+'_logR.png',dpi=300)
    plt.show()
```
